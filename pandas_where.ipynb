```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.DataFrame({
    'Hourly_Salary': ['500.00', '10000.00', '200.00', '20.00', np.nan]
})
```

```python
df["Hourly_Salary"] = df["Hourly_Salary"].astype(float)
```

```python
# Example 1 keep only values under 1000
filtered_df = df.where(df < 1000)
filtered_df
```

```python
# Example 2 keep under 1000, replace with another value
filtered_df = df.where(df < 1000, other = 999)
filtered_df
```

```python
# Example 3 filling a null value
df_where_3 = df.where(df.notnull(), 0)
df_where_3
```

```python
df2 = pd.DataFrame({
    'Running Back': ['Barry Sanders', 'Walter Payton', 'Emmitt Smith', 'Jim Brown'],
    'Career Rushing Yards': [152690, 16726, 18355, 12312],
    'Touchdowns': [99, 110, 164, 106]
})
```

```python
df2
```

```python
# Example 4 column
df2.where(df2["Career Rushing Yards"] < 100000)
```

```python
# Example 5 multiple conditions with &
df2.where( (df2["Touchdowns"] >= 99) & (df2["Career Rushing Yards"] != 18355) )
```

```python
# Example 6 multiple conditions with |
filter1 = df2["Touchdowns"] < 100
```

```python
filter2 = df2["Touchdowns"] > 150
```

```python
df2.where(filter1 | filter2)
```

```python
# Example 7 create a new column
df2["touchdown_totals"] = df2["Touchdowns"].where(df2["Touchdowns"] >= 100, other="Less Than 100")
```
