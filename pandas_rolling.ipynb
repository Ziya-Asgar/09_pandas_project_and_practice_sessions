```python
# Example 1 rolling avg last 3
```

```python
df["rolling_mean_3"] = df["Capacity"].rolling(window=3).mean()
```

```python
df.head(10)
```

```python
# Example 2 rolling sum
```

```python
df["rolling_sum"] = df["Capacity"].rolling(window=3).sum()
```

```python
df.head(10)
```

```python
# Example 3 Min/Max
```

```python
df["rolling_min"] = df["Capacity"].rolling(window=3).min()
df["rolling_max"] = df["Capacity"].rolling(window=3).max()
```

```python
df.head(10)
```

```python
# Example 4 rolling window different columns
```

```python
np.random.seed(42)
```

```python
df["tickets_sold"] = (df["Capacity"] * np.random.uniform(0.9, 1.0, len(df))).astype(int)
```

```python
df.head()
```

```python
df["tickets_sold"].unique()
```

```python
df[["Rolling_mean_Capacity_5", "Rolling_mean_Tickets_5"]] =\
  df[["Capacity", "tickets_sold"]]\
  .rolling(window=5).mean()
```

```python
df.head(10)
```

```python
# Example 5
```

```python
rolling_stats = df["Capacity"].rolling(window=5).agg(["mean", "sum", "std"])
```

```python
rolling_stats.head(10)
```

```python
# Example 6 use apply
```

```python
df["rolling_range"] = df["tickets_sold"].rolling(window=3).apply(lambda x: x.max() - x.min())
```

```python
df[["tickets_sold", "rolling_range"]].head(10)
```

```python
# Example 7 groupby
```

```python
# find the mean tickets sold for every 2-window occurrence of a venue
df["Rolling_mean_venue"] = df.groupby("Venue")["tickets_sold"]\
  .rolling(window=2).mean()\
  .reset_index(level=0, drop=True)
```

```python
df[["Venue", "Rolling_mean_venue"]].head(20)
```

```python
# Example 8 min periods
```

```python
df["rolling_avg_min_periods"] = df["Capacity"].rolling(window=5, min_periods=3).mean()
```

```python
df["rolling_avg_min_periods"].head(10)
```

```python
# Example use shift with rolling
```

```python
df["rolling_shift"] = df["Capacity"].rolling(window=5).mean().shift(1) - df["Capacity"]
```

```python
df["rolling_shift"].head(10)
```

```python
# here is what's going on when we use `.shift`
pd.DataFrame([
  df["Capacity"],
  df["Capacity"].shift(1), 
  df["Capacity"].rolling(window=5).mean(),
  df["Capacity"].rolling(window=5).mean().shift(1), 
  df["Capacity"].rolling(window=5).mean().shift(1) - df["Capacity"]
], index=[
  "Capacity", 
  "Capacity_shifted", 
  "Rolling_mean_5", 
  "Rolling_mean_5_shifted", 
  "Rolling_mean_5_shifted_vs_Capacity"]
).T.head(10)
```

```python
# Example 10 handling null values ffill
```

```python
df["Rolling_avg_ffill"] = df["Capacity"].ffill().rolling(window=5).mean()
```
